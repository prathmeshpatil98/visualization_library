```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
x = np.linspace(1,10,200)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x,y ,'--r')
plt.xlabel("this my x data")
plt.ylabel("this my y data")
plt.title("My sine wave chart")
plt.grid()
plt.show
```

```python
plt.figure(figsize = (12,9))
sns.barplot(x = x,
           y = y,
           palette = 'dark:salmon_r')
#plt.xticks(rotation = 90)
plt.xlabel("this my X data")
plt.ylabel("this is my y data")
plt.title('Total Attack')
plt.show()
```

```python
x1 = np.random.rand(50)
y1 = np.random.rand(50)
x1
```

```python
plt.scatter(x1,y1)
plt.xlabel("this my X data")
plt.ylabel("this is my y data")
plt.title('Total Attack')
plt.show()
```

```python
colours = np.random.rand(50)
sizes = 1000 * np.random.rand(50)
plt.figure(figsize=(17,6))
plt.scatter(x1,y1,c = colours,s= sizes, alpha=0.8)
plt.xlabel("this my X data")
plt.ylabel("this is my y data")
plt.title('graph')
#plt.xticks(rotation= 270)
plt.show()
```

```python
x = ('a','b' ,'c','d','e')
y = np.random.rand(5)
plt.subplots(figsize=(10,5))
plt.bar(x,y)
plt.xlabel("this my X data")
plt.ylabel("this is my y data")
plt.title('Practice Bar chart')
#plt.xticks(rotation= 270)
plt.show()
```

```python
x = ('a','b' ,'c','d','e')
y = np.random.rand(5)
plt.subplots(figsize=(18,9))
plt.barh(x,y)
plt.xlabel("this my X data")
plt.ylabel("this is my y data")
plt.title('Practice Bar chart')
#plt.xticks(rotation= 270)
plt.grid()
plt.show()
```

```python
colours = np.random.rand(50)
sizes = 1000 * np.random.rand(50)
plt.figure(figsize=(17,10))
plt.scatter(x1,y1,c = colours,s= sizes, alpha=0.8,marker='*')
plt.xlabel("this my X data")
plt.ylabel("this is my y data")
plt.title('my scatter graph')
#plt.xticks(rotation= 270)
plt.show()
```
